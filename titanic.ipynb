# TERCER GRAFICO

```python
import numpy as np

datos = np.array([10,40,20,30]) # Datos del grafico
etiquetas = ["Brownie", "Galletitas", "Cocadas", "Torta"] # Nombre de los numeros
explotar = [0,0.2,0,0] # Remarcar 1 parte del grafico

plt.pie(datos, labels=etiquetas, explode=explotar ,autopct="%.2f%%") #dejar en %

plt.show()
```
